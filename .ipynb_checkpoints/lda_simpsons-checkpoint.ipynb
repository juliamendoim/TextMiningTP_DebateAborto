```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import gensim
import numpy as np
import spacy

from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel
from gensim.models.wrappers import LdaMallet
from gensim.corpora import Dictionary
from gensim import corpora

import pyLDAvis.gensim
pyLDAvis.enable_notebook()

import os, re, operator, warnings
warnings.filterwarnings('ignore')  # Let's not pay heed to them right now
%matplotlib inline
```

```python
df = pd.read_csv("simpsons_script_lines.csv", error_bad_lines=False)
df.head()
```

```python
held_out = df[df.episode_id > 500]
development_df = df[df.episode_id <= 500]
del(df)
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
from nltk.tokenize import RegexpTokenizer
from collections import Counter

def proc(x):
    tokenizer = RegexpTokenizer(r'\w+')
    return [word.lower() for word in tokenizer.tokenize(x) if word not in STOP_WORDS]
    
def pre_process_corpus(df):
    all_speech = df.dropna(subset=['normalized_text']).groupby('episode_id')['normalized_text'].agg(' '.join)

    STOP_WORDS.add("im")
    STOP_WORDS.add("oh")
    STOP_WORDS.add("dont")
    STOP_WORDS.add("like")

    words = [w for t in all_speech.values for w in t.split(" ")]
    counts = Counter(words)

    all_speech = pd.DataFrame(all_speech.apply(lambda x: proc(x)))
    all_speech.tail(10)
    texts = all_speech.normalized_text.values
    return texts

texts = pre_process_corpus(development_df)
texts[0][0:10]
```

```python
dictionary = corpora.Dictionary(texts)
print(len(dictionary))
dictionary.filter_extremes()
print(len(dictionary))
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
lsimodel = LsiModel(corpus=corpus, num_topics=10, id2word=dictionary)
```

```python
lsimodel.print_topics(10, 4)
```

```python
hdpmodel = HdpModel(corpus=corpus, id2word=dictionary)
```

```python
hdpmodel.print_topics(10, 4)
```

```python
ldamodel = LdaModel(corpus=corpus, num_topics=10, id2word=dictionary, iterations=2000, passes=10)
```

```python
ldamodel.print_topics(10, 4)
```

```python

new_text = """

Tell you what, sir, from now on, you'll be Homer Thompson at Terror Lake. Let's just practice a bit, hmm? When I say "hello, Mr. Thompson", you'll say "hi".
Check!
Hello, Mr. Thompson.
Remember now, your name is Homer Thompson.
I gotcha.
Hello, Mr. Thompson.
Argh... Now when I say "hello, Mr. Thompson" and press down on your foot, you smile and nod.
No problem.
Hello, Mr. Thompson.
I think he's talking to you.
"""

# transform text into the bag-of-words space
bow_vector = dictionary.doc2bow(proc(new_text))
print([(dictionary[id], count) for id, count in bow_vector])
lda_vector = ldamodel[bow_vector]
print(lda_vector)
# print the document's single most prominent LDA topic
print(ldamodel.print_topic(max(lda_vector, key=lambda item: item[1])[0]))
```

```python
print(ldamodel)
```

```python
pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```

```python
pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary, mds="tsne")
```

```python
lsitopics  = [[word for word, prob in topic] for topicid, topic in lsimodel.show_topics(formatted=False)]
hdptopics = [[word for word, prob in topic] for topicid, topic in hdpmodel.show_topics(formatted=False)]
ldatopics = [[word for word, prob in topic] for topicid, topic in ldamodel.show_topics(formatted=False)]
```

```python
lsi_coherence = CoherenceModel(topics=lsitopics[:10], texts=texts, dictionary=dictionary, window_size=10).get_coherence()
hdp_coherence = CoherenceModel(topics=hdptopics[:10], texts=texts, dictionary=dictionary, window_size=10).get_coherence()
lda_coherence = CoherenceModel(topics=ldatopics, texts=texts, dictionary=dictionary, window_size=10).get_coherence()
```

```python
def evaluate_bar_graph(coherences, indices):
    """
    Function to plot bar graph.
    
    coherences: list of coherence values
    indices: Indices to be used to mark bars. Length of this and coherences should be equal.
    """
    assert len(coherences) == len(indices)
    n = len(coherences)
    x = np.arange(n)
    plt.bar(x, coherences, width=0.2, tick_label=indices, align='center')
    plt.xlabel('Models')
    plt.ylabel('Coherence Value')
```

```python
evaluate_bar_graph([lsi_coherence, hdp_coherence, lda_coherence],
                   ['LSI', 'HDP', 'LDA'])
```

#### Mejorar la separación en tópicos de LDA variando los parámetros, el número de tópicos y los stopwords utilizados 
Para ello, podrán utilizar la métrica comentada en clase que utiliza los documentos no utilizados para entrenamiento. Sobre cada documento se utiliza la mitad para estimar la proporción de tópicos, y la otra mitad para calcular la probabilidad de aparcición de las palabras:

```python
def intra_inter(model, test_docs, num_pairs=10000):
    # Cada documento de test es partido en 2 partes. 
    part1 = [model[dictionary.doc2bow(tokens[: len(tokens) // 2])] for tokens in test_docs]
    part2 = [model[dictionary.doc2bow(tokens[len(tokens) // 2 :])] for tokens in test_docs]
    
    print("Similaridad coseno promedio entre partes de un mismo documento (más es mejor):")
    print(np.mean([gensim.matutils.cossim(p1, p2) for p1, p2 in zip(part1, part2)]))

    random_pairs = np.random.randint(0, len(test_docs), size=(num_pairs, 2))
    print("Similaridad coseno promedio entre 10,000 partes de documentos al azar (menos es mejor):")    
    print(np.mean([gensim.matutils.cossim(part1[i[0]], part2[i[1]]) for i in random_pairs]))
```

```python
held_out_texts = pre_process_corpus(held_out)
intra_inter(ldamodel, held_out_texts)
```
